```
import numpy as np

import tensorflow as tf

import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
```

```
data = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = data.load_data()

frame = np.reshape(x_train[-2], (28,28))
print(y_train[-2])

plt.imshow(frame)
plt.title('Frames')
plt.show()
```

```
x_train[-2].shape
```

```
print(y_train.shape)
print(x_train.shape)
```

```
x_train =  x_train / 255.0
x_test  =   x_test / 255.0
```

```
x_train[0]
```

```
model = Sequential()      # Start model

model.add(Flatten())      # Flatten the input into 1d array

model.add(Dense(128, activation='relu'))         # dense layer of 128 nodes, 
# model.add(Dropout(0.1))                              # dropout layer of 20% nodes neglected

model.add(Dense(10,  activation='softmax'))      # 10 nodes for 0-9 numbers detection,  softmax for distribution
                                                     # of probabilities of labels for 0-9
    

model.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy', metrics = ['accuracy']) #compile the model

model.fit(x_train, y_train, epochs = 10, validation_split = 0.2)             # Training the model
```

```
model.evaluate(x_test,y_test) 
```

```
indx = 1030

img = x_train[indx]

img = np.reshape(img , (1,28,28))

pred = model.predict(img)

print(np.argmax(pred))
plt.imshow(x_train[indx])
```
